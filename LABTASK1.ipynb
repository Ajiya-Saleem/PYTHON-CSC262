# LAB TASK 01: STUDENT GRADE CALCULATOR

```python
marks = int(input("ENTER MARKS BETWEEN 0 TO 100\t"))
if  marks>=91:
    print("GRADE A")
elif marks>=81:
    print("GRADE B")
elif marks>=71:
    print("GRADE C")
elif marks>=61:
    print("GRADE D")
elif marks>=50:
    print("GRADE E")
else:
    print("GRADE F")
```

# LAB TASK 02: FACTORIAL CALCULATOR

```python
num = int(input("ENTER A NUMBER TO FIND ITS FACTORIAL\n"))
factorial = 1
if num<0:
    print("FACTORIAL CAN'T BE OF A NEGATIVE NUMBER")
elif num==0:
    print("FACTORIAL OF 0 IS 1")
else:
    for i in range(1,num+1):
       factorial = factorial * i
    print(f"THE FACTORIAL OF {num} IS {factorial}")
```

# LAB TAST 03: FIBONACCI SERIES

```python
n_terms = int(input("ENTER THE NUM TERM UPTO WHICH YOU WANT A FIBONACCI SERIES "))
n1,n2 = 0,1
count = 0
if n_terms<=0:
    print("ENTER A POSITIVE NUMBER")
elif n_terms==1:
    print(f"FIBONACCI SEQUENCE UPTO {n_terms} : {n1}")
else:
    print("FIBONACCI SERIES")
    while count<=n_terms:
        print(n1)
        nth = n1+n2
        n1 = n2
        n2 = nth
        count+=1

```
