```python
def printParents(node, adj, parent):
    if (parent == 0):
        print(node, "-> Root")
    else:
        print(node, "->", parent)

    for cur in adj[node]:
        if (cur != parent):
            printParents(cur, adj, node)

def printChildren(Root, adj):
    q = []
    q.append(Root)
    vis = [0] * len(adj)

    while (len(q) > 0):
        node = q[0]
        q.pop(0)
        vis[node] = 1
        print(node, "-> ", end=" ")

        for cur in adj[node]:
            if (vis[cur] == 0):
                print(cur, " ", end=" ")
                q.append(cur)
        print()

def printLeafNodes(Root, adj):
    for i in range(0, len(adj)):
        if (len(adj[i]) == 1 and i != Root):
            print(i, end=" ")
    print()

def printDegrees(Root, adj):
    for i in range(1, len(adj)):
        print(i, ": ", end=" ")
        if (i == Root):
            print(len(adj[i]))
        else:
            print(len(adj[i]) - 1)

N = 7
Root = 1
adj = [[] for _ in range(N + 1)]

adj[1].append(2); adj[2].append(1)
adj[1].append(3); adj[3].append(1)
adj[1].append(4); adj[4].append(1)
adj[2].append(5); adj[5].append(2)
adj[2].append(6); adj[6].append(2)
adj[4].append(7); adj[7].append(4)

print("The parents of each node are:")
printParents(Root, adj, 0)

print("\nThe children of each node are:")
printChildren(Root, adj)

print("\nThe leaf nodes of the tree are:")
printLeafNodes(Root, adj)

print("\nThe degrees of each node are:")
printDegrees(Root, adj)
```

When I start at 'A', the order is: A B E C D F G

When I change the start to 'G', the order becomes: G C A F B E D

**Observation**: BFS always explores all neighbors at the current level before moving deeper. By changing the start node, the "levels" of the graph are recalculated from that new starting point.

```python
class Node:
    def __init__(self, state, parent, actions, total_cost):
        self.state = state
        self.parent = parent
        self.actions = actions
        self.total_cost = total_cost

# Creating the graph from the image
# Each action is a tuple: (Neighbor Name, Weight)
graph = {
    'A': Node('A', None, [('B', 6), ('C', 9), ('E', 1)], 0),
    'B': Node('B', None, [('A', 6), ('D', 3), ('E', 4)], 0),
    'C': Node('C', None, [('A', 9), ('F', 2), ('G', 3)], 0),
    'D': Node('D', None, [('B', 3), ('E', 5), ('F', 7)], 0),
    'E': Node('E', None, [('A', 1), ('B', 4), ('D', 5), ('F', 6)], 0),
    'F': Node('F', None, [('C', 2), ('E', 6), ('D', 7)], 0),
    'G': Node('G', None, [('C', 3)], 0)
}

def bfs_weighted(graph, start, goal):
    visited = []
    queue = []

    # Start node: total_cost is 0
    queue.append(graph[start])
    visited.append(start)

    while queue:
        current_node = queue.pop(0)

        # If we reached the goal, stop and return the cost
        if current_node.state == goal:
            return current_node.total_cost

        # Look at neighbors (neighbor_state, weight)
        for neighbor_state, weight in current_node.actions:
            if neighbor_state not in visited:
                visited.append(neighbor_state)

                # Update the cost: current path cost + edge weight
                new_cost = current_node.total_cost + weight

                # Create a temporary node object to track cost and parent
                neighbor_node = Node(neighbor_state, current_node, graph[neighbor_state].actions, new_cost)
                queue.append(neighbor_node)

    return "Goal not reachable"

# Run the task: Find cost from A to G
start_state = 'A'
goal_state = 'G'
cost = bfs_weighted(graph, start_state, goal_state)

print(f"BFS Path Cost from {start_state} to {goal_state} is: {cost}")
```

```python
class Node:
    def __init__(self, state, parent, actions, total_cost):
        self.state = state
        self.parent = parent
        self.actions = actions
        self.total_cost = total_cost

# The Graph dictionary based on the Romania Map image
graph = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Giurgiu': [('Bucharest', 90)],
    'Urziceni': [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
    'Vaslui': [('Urziceni', 142), ('Iasi', 92)],
    'Iasi': [('Vaslui', 92), ('Neamt', 87)],
    'Neamt': [('Iasi', 87)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
    'Eforie': [('Hirsova', 86)]
}

def bfs_pathfinder(graph, start, goal):
    visited = []
    queue = []

    # Create start node (State, Parent, Actions, Cost)
    start_node = Node(start, None, graph[start], 0)
    queue.append(start_node)
    visited.append(start)

    while queue:
        current_node = queue.pop(0)

        if current_node.state == goal:
            # Reconstruct the path by looking at parents
            path = []
            temp = current_node
            while temp:
                path.append(temp.state)
                temp = temp.parent
            return path[::-1], current_node.total_cost # Return reversed path and cost

        for neighbor, weight in graph[current_node.state]:
            if neighbor not in visited:
                visited.append(neighbor)
                new_cost = current_node.total_cost + weight
                new_node = Node(neighbor, current_node, graph[neighbor], new_cost)
                queue.append(new_node)
    return None, 0

# Execute
path, total_distance = bfs_pathfinder(graph, 'Arad', 'Bucharest')

print("Following is the BFS result for Lab Task 2:")
print("Path taken:", " -> ".join(path))
print("Total Distance:", total_distance)
```
