**Task 7 & 8: Structured Display for Multiple Instance**

```python
class Student:
    def display(self):
        # 'self' refers to whichever student is calling the function
        for key, value in self.__dict__.items():
            print(f"{key}: {value}")

student1 = Student()
student1.name = "Ali"
student1.id = 10

student2 = Student()
student2.name = "Hamza"
student2.id = 20

student1.display()
print("-" * 10)
student2.display()
```

**Task 9 & 10: Memory Analysis and Optimization**

```python
import sys
class NormalStudent:
    def __init__(self, name):
        self.name = name


class OptimizedStudent:
    __slots__ = ['name']
    def __init__(self, name):
        self.name = name

n = NormalStudent("Ali")
o = OptimizedStudent("Ali")

print(f"Normal Size: {sys.getsizeof(n.__dict__)} bytes")
print(f"Optimized Size: Memory is saved because __dict__ doesn't exist!")
```

**Object-Oriented Design and Area Comparison**

**TRAPEZOID AND PARALELLOGRAM**




```python
# STEP1: DESIGN CLASSES
class Trapezoid:
    def __init__(self, side_a, side_b, height):
        self.side_a = side_a
        self.side_b = side_b
        self.height = height

    def calculate_area(self):
        return ((self.side_a + self.side_b) / 2) * self.height

class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return self.base * self.height
# STEP2 & 3: Comparison Mechanism and Dictionary Storage
# Create objects
trap = Trapezoid(10, 20, 5) # (10+20)/2 * 5 = 75
para = Parallelogram(15, 6) # 15 * 6 = 90

areas_dict = {
    "Trapezoid": trap.calculate_area(),
    "Parallelogram": para.calculate_area()
}
print(f"Stored Areas: {areas_dict}")

#Step 4: Iterate Over Multiple Dimensions
heights = [5, 10, 15]
base = 10

print("Analyzing Parallelogram Variations:")
for h in heights:
    temp_para = Parallelogram(base, h)
    print(f"Base: {base}, Height: {h} -> Area: {temp_para.calculate_area()}")

#Step 5: Determine and Display the Maximum Area
max_shape = max(areas_dict, key=areas_dict.get)
max_value = areas_dict[max_shape]

print("-" * 20)
print(f"The shape with the maximum area is: {max_shape}")
print(f"Area value: {max_value}")
```
